Life Expectancy using Linear Regression

Linear Regression is a statistical model used to model the relationship between a dependent variable and one or more independent variables. It assumes that the relationship between the variables is linear, i.e., the change in the dependent variable is proportional in the change of the independent variable.


Linear Regression involves fitting a straight line to a set of data points in order to estimate the value of the dependent variable for a given value of the independent variable. The line is determined by minimizing the sum of the squared difference between the observed values of the dependent variable and the values predicted by the line.



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
url = 'https://raw.githubusercontent.com/risitadas/Machine-Learning-Projects-using-Python/main/Linear%20Regression/Life%20Expectancy%20Data.csv?token=GHSAT0AAAAAACA33L7GED7NJV5DCVOKHW6OZCADFUQ'
data = pd.read_csv(url)
data.head()
```

```python
data = data.fillna(0)
```

```python
data['Status'] = data['Status'].map({'Developing':0, 'Developed':1})
```

```python
data.info()
```

```python
data.describe()
```

```python
data.columns
```

```python
X = data[['Country', 'Year', 'Status', 'Adult Mortality',
       'infant deaths', 'Alcohol', 'percentage expenditure', 'Hepatitis B',
       'Measles ', ' BMI ', 'under-five deaths ', 'Polio', 'Total expenditure',
       'Diphtheria ', ' HIV/AIDS', 'GDP', 'Population',
       ' thinness  1-19 years', ' thinness 5-9 years',
       'Income composition of resources', 'Schooling']]

y = data['Life expectancy ']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
print(lr.intercept_)
```

```python
coeff_data = pd.DataFrame(lr.coef_,X.columns,columns=['Coefficient'])
coeff_data
```

```python
predictions = lr.predict(X_test)
```

```python
plt.scatter(y_test,predictions)
```

```python
from sklearn import metrics
print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```
